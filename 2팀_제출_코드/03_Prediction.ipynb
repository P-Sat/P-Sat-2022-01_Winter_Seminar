# 1. 데이터 불러오기 및 라이브러리 설치

```python
#드라이브 마운트
from google.colab import drive
drive.mount('/content/drive')

#라이브러리 설치
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.metrics import confusion_matrix, f1_score, roc_auc_score,roc_curve
import warnings
warnings.filterwarnings('ignore')
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt

#시간 측정
import time
tic=time.time()

#데이터 불러오기
train= pd.read_csv('/content/drive/MyDrive/train.csv', encoding = 'utf-8')
test=pd.read_csv('/content/drive/MyDrive/test.csv', encoding = 'utf-8')

#Train data의 feature 변수와 target 변수 따로 저장
x_data = train.drop(['target'],axis=1)
y_data = train['target']

#데이터 분할
#재현성을 위해 random_state=2022로 고정
X_train, X_test, Y_train, Y_test = train_test_split(x_data, y_data, test_size=0.25, random_state=2022)
print(X_train.shape, X_test.shape)
```

# 2. 데이터 스케일링 및 모델 학습

```python
# KNN 모델 사용시 거리 계산을 위해서 각 특성들을 스케일링 
scaler = StandardScaler() 
scaler.fit(X_train) 
X_train = scaler.transform(X_train) 
X_test = scaler.transform(X_test)

#KNN 모델 사용
#classifier accuracy plot으로 찾은 최적의 k값과 파라미터로 모델 생성
classifier = KNeighborsClassifier(n_neighbors=7, weights="distance", metric="euclidean")

# 모델 학습
classifier.fit(X_train, Y_train)
```

# 3. 모델 적합

```python
y_pred = classifier.predict(test)
```

# 4. 최종 제출 파일 저장

```python
#예측값 데이터 프레임 생성
id = []
for i in range(0,16384) :
  id.append(int(i))

predict = pd.DataFrame({'id': id, 'target':y_pred})
predict

#csv로 저장하기
predict.to_csv('team2_submission.csv', index = False)
```
